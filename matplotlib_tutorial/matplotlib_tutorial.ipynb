# Python Matplotlib Tutorial for Beginners

## What is Matplotlib?

- Matplotlib is a 2D python plotting library (similar interface with Matlab)
    - Visulize the data
    - Identify Trends
    - Identify outliers
    
    
- Compatible with both NumPy and Pandas for plotting

## Importing / Installing packages



```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

## 1. Line Graphs using Matplotlib & Numpy

```python
# Creating an array

a = np.array([1,2,3,4,5])
a

plt.plot(a)
plt.show()
```

```python
# Creating 2 arrays
a = np.array([1,2,3,4,5])
b = np.array([2,2,2,2,2])

plt.plot(a,b)
plt.show()
```

```python
##### Having more than 1 line - Adding Title, legends and x/y tickers

# creating the raw
a = np.array([1,2,3,4,5])
b = np.array([2,2,2,2,2])
c = np.array([3,2,4,6,5])

# increase size
plt.figure(figsize=(15,10))

plt.plot(a, color = 'green', linestyle = '-.', label="Rent FY20")
plt.plot(b, linewidth=5., label="AVG Rent")
plt.plot(c, color = 'red', linestyle = '--', label="Rent FY19")
plt.title("Rent per Month Summary", fontsize=20, fontweight="bold")
plt.legend(loc="upper left", fontsize=15)
plt.xticks(range(0,len(a)),('Jan','Feb','Mar','Apr','May'))
#plt.xticks([0,1,2,3,4])
plt.xlabel("Months")
plt.ylabel("Rent $")
plt.show()
```

## 2. Scatter Plots using Matplotlib & Numpy

```python
# Simple Example

# np.arange: from 0 to the number you specify
# np.random.randint: From -> To ,Number of numbers you need
# np.random.randn(): random numbers with normal distribution (mean 0 and variance 1)

a = np.arange(30)
b = np.random.randint(0,30,30)

plt.scatter(a,b)
plt.show()
```

```python
# Example 2 - Scatter with size

s = np.random.randint(0,100,30)

b = a + np.random.randint(0,30,30)

plt.scatter(a,b, c="green", s=s)
plt.show()
```

## 3. Bar Plots using Matplotlib & Numpy

```python
# Example 1

# Creating the raw data
category = ['Account A','Account B','Account C']
revenue = np.array([300, 500, 700])


# Plotting it
plt.figure(figsize=(10,10))
plt.bar(range(0,3), revenue)
plt.xticks(range(0,3),category)
plt.show()
```

```python
# Example 2 - Subplots

# Creating the raw data
category = ['Account A','Account B','Account C']
revenue = np.array([300, 450, 700])

# Plotting it
plt.figure(figsize=(10,5))

# bar
plt.subplot(131)
plt.bar(range(0,3), revenue)
plt.xticks(range(0,3),category)


# line
plt.subplot(132)
plt.plot(range(0,3), revenue, c="red")
plt.xticks(range(0,3),category)


# scatter
plt.subplot(133)
plt.scatter(range(0,3), revenue, c="green")
plt.xticks(range(0,3),category)
plt.show()
```

## 5. Lineplots using Pandas' DFs

```python
# Loading the data

raw_data = pd.read_csv("https://raw.githubusercontent.com/rahulg-infiswift/AnalyticsCourse/main/datasets/MarketingRawData.csv")
# runs all the data
raw_data

print(raw_data.shape)

#runs the first 5 rows
raw_data.head()
```

```python
# Example 1 - Simple line graph

plt.plot(raw_data["Visitors"])
plt.show()
```

```python
# Example 2 - Line graphs with xtickers

plt.plot(raw_data["Visitors"])
plt.xticks(range(0,len(raw_data["Visitors"])), raw_data["Date"], rotation = 90)
plt.show()
```

### Convert date into pandas datetime format 

```python
raw_data['Date'] = pd.to_datetime(raw_data['Date'], format = '%d/%m/%Y')
```

```python
raw_data.head(20)
```

```python
plt.plot(raw_data['Date'], raw_data["Visitors"])
plt.show()
```

```python
# Example 3 - Line graphs with xtickers and less data

a = raw_data.tail(10).reset_index()

plt.plot(a["Visitors"])
plt.xticks(range(0,len(a["Visitors"])), a["Date"], rotation = 90)
plt.show()
```

```python
# Example 4 - 2+ data points on the graph

plt.figure(figsize=(15,10))
plt.plot(a["Visitors"], label="No. of Visistors", linewidth=5.)
plt.plot(a["Revenue"], linestyle = "-.")
plt.xticks(range(0,len(a["Visitors"])), a["Date"], rotation = 90)
plt.title("Visitors & Revenue by Day", fontsize = 20)
plt.legend(loc="upper right", fontsize=14)
plt.xlabel("Dates")
plt.show()
```

```python
# Example 5 - Combo graphs

plt.figure(figsize=(15,10))
plt.bar(range(0,len(a["Visitors"])), a["Visitors"], label="No. of Visitors", linewidth=5., color = "green")
plt.plot(a["Revenue"], linestyle = "-.")
plt.xticks(range(0,len(a["Visitors"])), a["Date"], rotation = 90)
plt.title("Visitors & Revenue by Day", fontsize = 20)
plt.legend(loc="upper right", fontsize=14)
plt.xlabel("Dates")
plt.show()
```

```python
# Example 6 - Combo stack bars


plt.figure(figsize=(15,10))
plt.bar(range(0,len(a["Revenue"])), a["Revenue"], label="Revenue", linewidth=5., color = "green")
plt.bar(range(0,len(a["Visitors"])), a["Marketing Spend"], label="Marketing Spend", linewidth=5., color = "red")

plt.plot(a["Visitors"], linestyle = "-.")

plt.xticks(range(0,len(a["Visitors"])), a["Date"], rotation = 90)
plt.title("Visitors & Revenue by Day", fontsize = 20)
plt.legend(loc="upper right", fontsize=14)
plt.xlabel("Dates")
plt.show()

```

```python
# Example 7 - Combo 2+ bar graphs + Secondary Scale

weight = .30

a[["Revenue", "Marketing Spend"]].plot(kind="bar", color = ["green", "red"], figsize=(15,10))
a["Visitors"].plot(c="blue", linestyle = "-.", secondary_y=True)

plt.xticks(range(0,len(a["Visitors"])), a["Date"], rotation = 90)
plt.title("Visitors & Revenue by Day", fontsize = 20)
plt.legend(loc="upper left", fontsize=14)
plt.xlabel("Dates")
plt.xlim(-weight, len(a["Visitors"])-weight)
plt.show()

# examples of Kind
# https://pandas.pydata.org/pandas-docs/stable/user_guide/visualization.html
```

## 6. Scatterplot Using DFs

```python
# Example 1 - Simple Scatterplot

plt.figure(figsize=(15,10))
plt.scatter(raw_data["Revenue"], raw_data["Marketing Spend"], c="red", marker='+')
plt.title("Relationship between M. Spend & Revenue", fontsize = 20)
plt.legend(loc="upper left", fontsize=14)
plt.xlabel("Revenue")
plt.ylabel("Marketing Spend")

plt.show()
```

```python
raw_data.head()

```

```python
# Example 2 - Scatterplot with groups

groups = raw_data[["Promo","Revenue","Marketing Spend"]].groupby("Promo")
groups

fig, ax = plt.subplots(figsize=(15,10))

for promo, group in groups:
    ax.scatter(group['Revenue'], group['Marketing Spend'], marker = 'x', label=promo)

ax.legend()
plt.title("Relationship between M. Spend & Revenue by Promo", fontsize = 20)
plt.xlabel("Revenue")
plt.ylabel("Marketing Spend")

plt.show()
```

```python
# Example 3 - Scatterplot with groups

groups = raw_data[["Day_Name","Revenue","Marketing Spend"]].groupby("Day_Name")
groups

fig, ax = plt.subplots(figsize=(15,10))

for promo, group in groups:
    ax.scatter(group['Revenue'], group['Marketing Spend'], marker = 'x', label=promo)

ax.legend()
plt.title("Relationship between M. Spend & Revenue by Day_Name", fontsize = 20)
plt.xlabel("Revenue")
plt.ylabel("Marketing Spend")

plt.show()
```
