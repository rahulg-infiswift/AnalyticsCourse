```python
df['closure_time'] = (df['Close Date/Expected Close Date'] - df['Created Date']).dt.days
```

```python
df
```

```python
df.columns
```

```python
df['business_value_cr'] = df['Potential Business'].replace('50L', 0.50).replace('1Cr', 1).replace('1-4Cr', 2.5).replace('1-4CR', 2.5).replace('4-7 CR', 5.5)
```

```python
df.head() 
```

```python
df['expected_business_cr'] = df.business_value_cr*df[' Win Probability']
```

```python
df.head()
```

```python
df['Created Date'].min(), df['Created Date'].max()
```

```python
# sale_channel = p * expected_business_cr
# sale_channel = q * 1/closure_time
```

```python
df['cr_per_day'] = df.expected_business_cr/df.closure_time
```

```python
df.groupby(['Sales stage']).agg({' Win Probability':'mean'})
```
